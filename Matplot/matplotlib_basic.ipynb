```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

## 1 Plotting

```python
# compute the x, y coordinates for points on a sin curve
import numpy as np

x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y)
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)

plt.xlabel("x axis label")
plt.ylabel("y axis label")

plt.title("Sine and Cosine")
plt.legend(['Sine', 'Cosine'])
```

## 2 Subplots

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)

plt.plot(x, y_sin)
plt.title("Sine")

plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title("Cosine")

plt.show()
```

## 3 Line plots

```python
xvals = np.arange(-2, 1, 0.01)
yvals = np.cos(xvals)
plt.plot(xvals, yvals)
plt.show()
```

```python
xvals = np.arange(-2, 1, 0.01)
yvals = np.cos(xvals)
plt.plot(xvals, yvals)

newyvals = 1 - 0.5 * xvals ** 2
plt.plot(xvals, newyvals, 'r--')
plt.title("Example plots")
plt.xlabel("Input")
plt.ylabel("Function values")
plt.show()
```

## 4 Contour plot

```python
plt.figure()
xlist = np.linspace(-2.0, 1.0, 100)
ylist = np.linspace(-1.0, 2.0, 100)

X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2+Y**2)

plt.contour(X, Y, Z, [0.5, 1.0, 1.2, 1.5], colors = 'k', linestyles = 'solid')
```

```python
plt.contour(X, Y, Z, [0.5, 1.0, 1.2, 1.5], colors = 'k', linestyles = 'solid')
plt.axes().set_aspect('equal')
```
